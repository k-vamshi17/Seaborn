```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,6))
```

```python
# 1. scatter plot

sns.scatterplot(data=tips,x="total_bill", y="tip", hue="time", size="size",palette="deep")
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
# 2. Line Plot
sns.lineplot(data=tips, x='size', y='total_bill', hue='sex', markers='0' )
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
sns.lineplot(data=tips, x='size', y='total_bill', hue='sex', markers='0', ci=None )
plt.title("Lineplot of Total Bill vs Size")
plt.show()
```

```python
#3. bar plot
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', palette='muted')
plt.title("Barplot of Total Bill")
plt.show()
```

```python
#4. boxplot
sns.boxplot(data=tips, x='day', y='tip', hue='smoker' , palette='bright')
plt.show("Box of Tips by Day ans Smoker Status")
plt.show()
```

```python
#5. Violen Plot
sns.violinplot(data=tips, x='day' , y='total_bill', hue='time', palette='pastel', split=True)
plt.title("Boxplot of Tips by Day and Smoker Status")
plt.show()
```

```python
#6 count plot
sns.countplot(data=tips, x='day', hue="smoker", palette='dark')
plt.title("Count Plot of Days by Smoker Status")
plt.show()
```

```python
#7. regression plot
sns.regplot(data=tips, x='total_bill',y='tip', scatter_kws={'s':50}, line_kws={'color':'red'} )
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

```python
#8. Histogram
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='blue')
plt.title("Histogram of Total Bill")
plt.show()
```

```python
#9. pairplot
sns.pairplot(tips, hue='sex', vars=['total_bill', 'tip', "size"] , palette='husl')
plt.suptitle("pair plot: numberic variables by gender", y=1.02)
plt.show()
```

```python
#10. catplot
sns.catplot(data=tips, x='day', y='tip', hue='sex', kind='point', palette='bright')
plt.suptitle("catplot(point): Tips by day and gender")
plt.show()
```

```python
#11. jointplot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker',color='purple', palette='pastel')
plt.suptitle("jointplot: Total Bill vs Tip by Smoker Status", y=1.02)
plt.show()
```

```python
#12. FacetGrid
sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.scatterplot, 'total_bill', 'tip', color='purple')
plt.suptitle("FacetGrid: Total Bill vs Tip by Smoker Status", y=1.02)
plt.show()
```

```python
#13. stripplot
sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True, palette='deep')
plt.title("Stripplot of Total Bill by Day and Smoker Status")   
plt.show()
```

```python
#14. KDE plot
sns.kdeplot(data=tips, x='total_bill', hue='sex',  fill=True, palette='tab10')  
plt.title("KDE plot: Total Bill density and gender")
plt.show()
```
